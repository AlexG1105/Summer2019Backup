```python
#Now, we attempt to count the set of commonly occuring groups of appliances in both eval and train sets
countDict = {}
for i, row in enumerate(df4.itertuples()):
    current = json.loads(row.annotation)
    listItem = []
    for j, key in enumerate(current):
        for classDict in key['class']:
            listItem.append(classDict)
    addSet = frozenset(set(listItem))
    if addSet in countDict.keys():
        countDict[addSet] += 1
    else:
        countDict[addSet] = 1
```

```python
df1exp = pd.DataFrame()

for i,row in enumerate(df1.itertuples()):
    removeidx = []
    current = json.loads(row.annotation)
    for j, key in enumerate(current):
        for classDict in key['class']:
            if classDict not in df1cat:
                removeidx.append(j)
    
    for index in sorted(removeidx, reverse=True):
        del current[index]
    
    for k, key in enumerate(current):
        temp = json.dumps(key)
        df1exp = pd.concat([df1exp, pd.DataFrame([[row[1], temp]])]).reset_index(drop = True)
        
images_path = 'applianceImages'
image_extensions = ['.jpg', '.png', '.jpeg']
images = [os.path.join(dp, f) for dp, dn, filenames in os.walk(images_path)
         for f in filenames if os.path.splitext(f)[1].lower() in image_extensions]
len(images)

imageIDs = [x.split('.')[0].split('/')[-1] for x in images]
imagepathlist1 = []
imagePair = list(zip(images,imageIDs))

for i,row in enumerate(df1exp.itertuples()):
    for pair in imagePair:
        if str(row[1]) in pair[1]:
            imagepathlist1.append(pair[0])
            continue
            
df1exp['filename'] = pd.DataFrame(imagepathlist1)
```

```python
xmldf1 = pd.DataFrame()

for i,row in enumerate(df1exp.itertuples()):
    with Image.open(row[3]) as im:
        w,h = im.size
    current = json.loads(row[2])
    for classDict in current['class']:
        tempClass = classDict
        
    for coor in current['coordinates']:
        if coor == 'h':
                temph = current['coordinates'][coor]
        elif coor == 'w':
            tempw = current['coordinates'][coor]
        elif coor == 'x':
            xmin = current['coordinates'][coor]
        elif coor == 'y':
            ymin = current['coordinates'][coor]

    xmax = xmin + tempw
    ymax = ymin + temph
    
    xmldf1 = pd.concat([xmldf1, pd.DataFrame([[row[3],w,h,tempClass,xmin,ymin,xmax,ymax]])]).reset_index(drop=True)
    
xmldf1.columns = ['filename','width','height','class','xmin','ymin','xmax','ymax']
```
