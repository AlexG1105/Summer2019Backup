```python
#Generate the counts for each csv, and with the top 5, we set those as the categories for the new object detection thing
cd1 = {}
cd2 = {}
cd3 = {}
cd4 = {}
cd5 = {}

for i,row in enumerate(df1filter.itertuples()):
    current = json.loads(row.annotation)
    for j,key in enumerate(current):
        for classDict in key['class']:
            if classDict in cd1.keys():
                cd1[classDict] += 1
            else:
                cd1[classDict] = 1
                
for i,row in enumerate(df2.itertuples()):
    current = json.loads(row.annotation)
    for j,key in enumerate(current):
        for classDict in key['class']:
            if classDict in cd2.keys():
                cd2[classDict] += 1
            else:
                cd2[classDict] = 1
                
for i,row in enumerate(df3filter.itertuples()):
    current = json.loads(row.annotation)
    for j,key in enumerate(current):
        for classDict in key['class']:
            if classDict in cd3.keys():
                cd3[classDict] += 1
            else:
                cd3[classDict] = 1
                
for i,row in enumerate(df4filter.itertuples()):
    current = json.loads(row.annotation)
    for j,key in enumerate(current):
        for classDict in key['class']:
            if classDict in cd4.keys():
                cd4[classDict] += 1
            else:
                cd4[classDict] = 1
                
for i,row in enumerate(df5.itertuples()):
    current = json.loads(row.annotation)
    for j,key in enumerate(current):
        for classDict in key['class']:
            if classDict in cd5.keys():
                cd5[classDict] += 1
            else:
                cd5[classDict] = 1
                
bathroomcounts = sorted(cd1.items(), key=operator.itemgetter(1), reverse = True)
outdoorcounts = sorted(cd2.items(), key=operator.itemgetter(1), reverse = True)
kitchencounts = sorted(cd3.items(), key=operator.itemgetter(1), reverse = True)
bedroomcounts = sorted(cd4.items(), key=operator.itemgetter(1), reverse = True)
livingroomcounts = sorted(cd5.items(), key=operator.itemgetter(1), reverse = True)
```
