```python
import os
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from keras.models import Sequential, load_model
```

```python
img_width, img_height = 150, 150
model_path = './models/model.h5'
model_weights_path = './models/weights.h5'
model = load_model(model_path)
model.load_weights(model_weights_path)
```

```python
def predict(file):
    x = load_img(file, target_size=(img_width,img_height))
    x = img_to_array(x)
    x = np.expand_dims(x, axis=0)
    array = model.predict(x)
    result = array[0]
    answer = np.argmax(result)
    if answer == 0:
        print("Label: Bathroom")
    elif answer == 1:
        print("Labels: Bedroom")
    elif answer == 2:
        print("Label: Dining Room")
    elif answer == 3:
        print("Label: Entryway")
    elif answer == 4:
        print("Label: Kitchen")
    elif answer == 5:
        print("Label: Living Room")
    elif answer == 6:
        print("Label: Office")
    elif answer == 7:
        print("Outdoor")

    return answer
```

```python
prediction = predict('filteredImages/bedroom/1954333630.jpg')
```
